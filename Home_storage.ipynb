```python
print('Witamy w najlepszej aplikacji do zarządzania domową spiżarnią')

import json

try:
    with open('warehouse.json', 'r') as fp:
        warehouse = json.load(fp)
    with open('limits.json', 'r') as fp:
        limits = json.load(fp)
        
except:
    warehouse = {}
    limits = {}
    
    with open('warehouse.json', 'w') as fp:
        json.dump(warehouse, fp)
        
    with open('limits.json', 'w') as fp:
        json.dump(limits, fp)
        

def save_warehouse(warehouse: dict):
    with open('warehouse.json', 'w') as fp:
        json.dump(warehouse, fp)


def save_limits(limits: dict):
    with open('limits.json', 'w') as fp:
        json.dump(limits, fp)


print('''Co chciałbys zrobić:
1 - dodać zakupione artykuły do spiżarni
2 - usunąć zużyte artykuły ze spiżarni
3 - dodać nowy artykuł
4 - usunąć artykuł
5 - zmienić limity
6 - sprawdzić stan magazynowy
7 - wydrukowac listę zakupów
9 - zakończyć działanie aplikacji''')

choice = 1

while choice != 8:
    choice = int(input('Co chciałbyś zrobić:' ))
    
    if choice==1:
        item = input('Podaj nazwę artykułu:').capitalize()
        while True:
            try:
                quantity = int(input('Podaj ilość:'))
                if quantity in range (1, 10000):
                    break
                print('That is not a positive number!')
            except:
                print('That is not a positive number!')
        warehouse[item] = warehouse.setdefault(item, 0) + int(quantity)
        save_warehouse(warehouse)

        if item not in limits:
            while True:
                try:
                    minimum = int(input('Podaj minimalną ilość:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[item] = int(minimum)
        save_limits(limits)
        
        
    elif int(choice) == 2:
        item = input('Podaj nazwę zużytego artykułu:').capitalize()
        if item in warehouse:
            while True:
                try:
                    quantity = int(input('Podaj zużytą ilość:'))
                    if quantity in range (1, 10000):
                        break
                    print('That is not a positive!')
                except:
                    print('That is not a positive number!')
            if warehouse[item] > 0:
                warehouse[item] -= quantity
            elif warehouse[item] <= 0:
                print('Nie można wykonac operacji. Za mała ilość towaru w magazynie')
            if warehouse[item] < 0:
                warehouse[item] = 0
            save_warehouse(warehouse)
        else:
            print('Brak towaru w bazie danych')       
        

    elif int(choice) == 3:
        new_item = input('Podaj nazwę nowego artykułu:').capitalize()
        if new_item not in warehouse:
            warehouse[new_item] = 0
            save_warehouse(warehouse)
        else:
            print('This product exist in warehouse.')
        if new_item not in limits:
            while True:
                try:
                    minimum = int(input('Podaj minimalną ilość:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[new_item] = int(minimum)
            save_limits(limits)
            print('You have added new product:', new_item)
    
    
    elif int(choice) == 4:
        del_item = input('Podaj nazwę artykułu do usunięcia:').capitalize()
        if del_item in warehouse:
            warehouse.pop(del_item)
            save_warehouse(warehouse)
            limits.pop(del_item)
            save_limits(limits)
            print('You have deleted the following product:', del_item)
        else:
            print('Nie ma takiego towaru')
    
    
    elif int(choice) == 5:
        limit_item = input('Podaj nazwę artykułu, którego chcesz zmienić limity:').capitalize()
        if limit_item in warehouse:
            while True:
                try:
                    new_minimum = int(input('Podaj nową minimalną ilość:'))
                    if new_minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[limit_item] = int(new_minimum)
            save_limits(limits)
            print('You have changed limits of the following product:', limit_item)
        else:
            print('Nie ma takiego artykułu')
     
    
    elif int(choice) == 6:
        print('Towar: ilość')
        for key, value in warehouse.items():
            print('{}: {} szt.'.format(key, value))
   

    elif int(choice) == 7:
        print('Należy zakupić:')
        for key, value in warehouse.items():
            if warehouse[key] <= limits[key]:
                print('{}: minimum {} szt.'.format(key, (limits[key]-warehouse[key])))

    
    elif int(choice) == 8:
        #print(warehouse, limits)
        break
        
    else:
        print(input('Co chciałbyś zrobić:' ))

print('Dzięki za skorzystanie z naszego programu.')
```
