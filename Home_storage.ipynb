```python
print('Witamy w najlepszej aplikacji do zarządzania domową spiżarnią')

warehouse = {}
limits = {}

print('''Co chciałbys zrobić:
1 - dodać zakupione artykuły do spiżarni
2 - usunąć zużyte artykuły ze spiżarni
3 - dodać nowy artykuł
4 - usunąć artykuł
5 - zmienić limity
6 - sprawdzić stan magazynowy
7 - wydrukowac listę zakupów
8 - zakończyć działanie aplikacji''')

choice = 1

while choice != 8:
    choice = int(input('Co chciałbyś zrobić:' ))
    
    if choice==1:
        item = input('Podaj nazwę artykułu:').capitalize()
        while True:
            try:
                quantity = int(input('Podaj ilość:'))
                if quantity in range (1, 10000):
                    break
                print('That is not a positive number!')
            except:
                print('That is not a positive number!')
        warehouse[item] = warehouse.setdefault(item, 0) + int(quantity)
        #save_warehouse(warehouse)

        if item not in limits:
            while True:
                try:
                    minimum = int(input('Podaj minimalną ilość:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[item] = int(minimum)
        #save_limits(limits)
        
        
    elif int(choice) == 2:
        item = input('Podaj nazwę zużytego artykułu:').capitalize()
        if item in warehouse:
            while True:
                try:
                    quantity = int(input('Podaj zużytą ilość:'))
                    if quantity in range (1, 10000):
                        break
                    print('That is not a positive!')
                except:
                    print('That is not a positive number!')
            if warehouse[item] > 0:
                warehouse[item] -= quantity
            elif warehouse[item] <= 0:
                print('Nie można wykonac operacji. Za mała ilość towaru w magazynie')
            if warehouse[item] < 0:
                warehouse[item] = 0
        else:
            print('Brak towaru w bazie danych')       
        

    elif int(choice) == 3:
        item = input('Podaj nazwę nowego artykułu:').capitalize()
        if item not in warehouse:
            warehouse[item] = 0
        if item not in limits:
            while True:
                try:
                    minimum = int(input('Podaj minimalną ilość:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[item] = int(minimum)
    
    
    elif int(choice) == 4:
        del_item = input('Podaj nazwę artykułu do usunięcia:').capitalize()
        if del_item in warehouse:
            warehouse.pop(del_item)
        else:
            print('Nie ma takiego towaru')
    
    
    elif int(choice) == 5:
        item = input('Podaj nazwę artykułu, którego chcesz zmienić limity:').capitalize()
        if item in warehouse:
            while True:
                try:
                    new_minimum = int(input('Podaj nową minimalną ilość:'))
                    if new_minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[item] = int(new_minimum)
        else:
            print('Nie ma takiego artykułu')
     
    
    elif int(choice) == 6:
        print('Towar: ilość')
        for key, value in warehouse.items():
            print('{}: {} szt.'.format(key, value))
   

    elif int(choice) == 7:
        print('Należy zakupić:')
        for key, value in warehouse.items():
            if warehouse[key] <= limits[key]:
                print('{}: minimum {} szt.'.format(key, (limits[key]-warehouse[key])))

    
    elif int(choice) == 8:
        break
        
    else:
        print(input('Co chciałbyś zrobić:' ))

print('Dzięki za skorzystanie z naszego programu.')
```
