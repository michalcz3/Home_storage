```python
print('Welcome to the best home pantry management application')

import json

try:
    with open('warehouse.json', 'r') as fp:
        warehouse = json.load(fp)
    with open('limits.json', 'r') as fp:
        limits = json.load(fp)
        
except:
    warehouse = {}
    limits = {}
    
    with open('warehouse.json', 'w') as fp:
        json.dump(warehouse, fp)
        
    with open('limits.json', 'w') as fp:
        json.dump(limits, fp)
        

def save_warehouse(warehouse: dict):
    with open('warehouse.json', 'w') as fp:
        json.dump(warehouse, fp)


def save_limits(limits: dict):
    with open('limits.json', 'w') as fp:
        json.dump(limits, fp)


print('''What would you like to do?
1 - add the purchased goods to the pantry
2 - remove the used goods from the pantry
3 - add the new good
4 - remove the good
5 - change the limits
6 - check the stock
7 - view the shopping list
8 - quit the application''')

choice = 1

while choice != 8:
    choice = int(input('What would you like to do?' ))
    
    if choice==1:
        item = input('Enter the name of the article:').capitalize()
        while True:
            try:
                quantity = int(input('Enter the quantity:'))
                if quantity in range (1, 10000):
                    break
                print('That is not a positive number!')
            except:
                print('That is not a positive number!')
        warehouse[item] = warehouse.setdefault(item, 0) + int(quantity)
        save_warehouse(warehouse)

        if item not in limits:
            while True:
                try:
                    minimum = int(input('Enter the minimum quantity:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[item] = int(minimum)
        save_limits(limits)
        
        
    elif int(choice) == 2:
        item = input('Enter the name of the used article:').capitalize()
        if item in warehouse:
            while True:
                try:
                    quantity = int(input('Enter the used quantity:'))
                    if quantity in range (1, 10000):
                        break
                    print('That is not a positive!')
                except:
                    print('That is not a positive number!')
            if warehouse[item] > 0:
                warehouse[item] -= quantity
            elif warehouse[item] <= 0:
                print('Error. Too few goods in the stock')
            if warehouse[item] < 0:
                warehouse[item] = 0
            save_warehouse(warehouse)
        else:
            print('No article in the database')       
        

    elif int(choice) == 3:
        new_item = input('Enter the name of the new good:').capitalize()
        if new_item not in warehouse:
            warehouse[new_item] = 0
            save_warehouse(warehouse)
        else:
            print('This product exist in pantry.')
        if new_item not in limits:
            while True:
                try:
                    minimum = int(input('Enter the minimum quantity:'))
                    if minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[new_item] = int(minimum)
            save_limits(limits)
            print('You have added new product:', new_item)
    
    
    elif int(choice) == 4:
        del_item = input('Enter the name of the good to be deleted:').capitalize()
        if del_item in warehouse:
            warehouse.pop(del_item)
            save_warehouse(warehouse)
            limits.pop(del_item)
            save_limits(limits)
            print('You have deleted the following product:', del_item)
        else:
            print('No article in the database')
    
    
    elif int(choice) == 5:
        limit_item = input('Enter the name of the article that you want to change the limits for:').capitalize()
        if limit_item in warehouse:
            while True:
                try:
                    new_minimum = int(input('Enter the new minimum quantity:'))
                    if new_minimum in range (1, 10000):
                        break
                    print('That is not a positive number!')
                except:
                    print('That is not a positive number!')                
            limits[limit_item] = int(new_minimum)
            save_limits(limits)
            print('You have changed limits of the following product:', limit_item)
        else:
            print('No article in the database')
     
    
    elif int(choice) == 6:
        print('Good: quantity')
        for key, value in warehouse.items():
            print('{}: {} pcs.'.format(key, value))
   

    elif int(choice) == 7:
        print('You should buy:')
        for key, value in warehouse.items():
            if warehouse[key] <= limits[key]:
                print('{}: minimum {} pcs.'.format(key, (limits[key]-warehouse[key])))

    
    elif int(choice) == 8:
        #print(warehouse, limits)
        break
        
    else:
        print(input('What would you like to do?' ))

print('Thanks for using application.')
```
